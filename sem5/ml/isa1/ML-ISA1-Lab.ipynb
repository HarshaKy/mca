# Import and Preprocessing

```python
import pandas as pd
import numpy as np
df = pd.read_csv('Redwine.csv', delimiter = ';')
df.head()
```

```python
df[:10]
```

```python
df.keys()
```

```python
x = df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']]
```

```python
y = df['quality']
y[:5]
```

# Split into train and test and StandardScaler

```python
from sklearn.model_selection import train_test_split

xTrain, xTest, yTrain, yTest =  train_test_split(x, y, test_size = 0.3, random_state=33) 
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(xTrain)
standardized_X = scaler.transform(xTrain)
standardized_X_test = scaler.transform(xTest)
```

# Build SVC

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(standardized_X, yTrain)
```

# Evaluation and Accuracy Score

```python
predictions = model.predict(standardized_X_test)
predictions
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(yTest, predictions)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(yTest, predictions))
```

```python
print(classification_report(yTest, predictions))
```

# GridSearch

```python
param_grid = {'C': [0.1, 1, 10, 100, 1000], 'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 'kernel': ['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3, cv = 9)
```

```python
grid.fit(xTrain, yTrain)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_predictions = grid.predict(xTest)
```

```python
print(confusion_matrix(yTest, grid_predictions))
```

```python
print(classification_report(yTest, grid_predictions))
```
